```python
import pandas as pd
```

```python
df1 = pd.read_excel(r"C:\Users\adeda\Downloads\Customer Call List.xlsx")
df1
```

```python
df1= df1.drop_duplicates()
df1
```

```python
df1 = df1.drop(columns = 'Not_Useful_Column')
df1
```

```python
df1['Last_Name'] = df1['Last_Name'].str.strip('123._/')
df1
```

```python
df1
```

```python
df1['Phone_Number'] = df1['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df1
```

```python
df1['Phone_Number'] = df1['Phone_Number'].astype(str)  # Convert to string

df1['Phone_Number'] = df1['Phone_Number'].apply(lambda x: x[:3] + '-' + x[3:6] + '-' + x[6:10])

df1
```

```python
df1['Phone_Number'] = df1['Phone_Number'].replace('nan--',' ')
df1['Phone_Number'] = df1['Phone_Number'].replace('Na--',' ')
df1
```

```python
df1[['Street_Address','State','Zip_code']] =  df1['Address'].str.split(',',2, expand = True)
df1
```

```python
df1['Paying Customer'] = df1['Paying Customer'].str.replace('Yes','Y')
df1['Paying Customer'] = df1['Paying Customer'].str.replace('No','N')
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace('Yes','Y')
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace('No','N')
df1
```

```python
df1 = df1.replace('N/a','')
df1
```

```python
df1 = df1.fillna('')
```

```python
for x in df1.index:
    if df1.loc[x,'Do_Not_Contact'] == 'Y':
        df1.drop(x, inplace = True)
df1

```

```python
for x in df1.index:
    if df1.loc[x,'Phone_Number'] == ' ':
       df1.drop(x, inplace = True)
df1

#another way to drop null value
#df1.dropna(subset='Phone_Number'), inplace = True
```

```python
df1.reset_index(drop=True)
```

```python
df1 = df1.drop(columns = 'Address')
```

```python
df1['Address'] = df1['Street_Address'] + ',' + df1['State']
df1
```
